```python
row_count = [0]

W_truck = 180 
L_truck = 420 
H_truck = 200 
H_lim = 200

product_information = []

# 시작점
position1_x = []
position1_y = []
position1_z = []
# 끝점
position2_x = []
position2_y = []
position2_z = []

truck_info = []
mod_row_lst = []
```

```python
# 품목 타입 & 타입 별 규격
product_type_index = ['A', 'B'] 
product_type = [[48,52,40], [38,48,34]]

# 데이터 불러오기
# window용 excel 불러오기
df_raw = pd.read_excel('C:/Users/user/Desktop/DATA/dataset.xlsx', sheet_name = 'initial_dataset')
df_raw.drop(df_raw.columns[0], axis=1, inplace=True)

print('__________DATASET DATAFRAME________')
display(df_raw)
```

```python
df_raw['S'] = 999

for i in range(len(df_raw)):
    df_raw['S'][i] = df_raw['L'][i] * df_raw['W'][i]

df_raw = df_raw[['L', 'W', 'H', 'S', 'destination', 'box_index', 'product_type']]

df_raw.sort_values(by=['destination', 'product_type'], ascending=[True, True], inplace=True)
df_raw.reset_index(inplace = True, drop = True)

print('____________df_raw__________')
display(df_raw)
```

```python
def preprocessing_A(df_raw, product_type_index, H_lim):
    
    df_block_std = pd.DataFrame() 
    df_block_data = pd.DataFrame() 
    
    for i in range(3):
        df_1 = df_raw[df_raw['destination'] == i+1] 
        type_order = product_type_index
        block_idx = 1
        
        for j in range(len(type_order)):
            df_2 = df_1[df_1['product_type'] == type_order[j]] 
            df_2.reset_index(inplace=True, drop=True)
            df_2['block_number'] = 999       
            cnt = 0 
            check_row = 0 
            check_height = df_2['H'][check_row] 
            
            while cnt < 1:
                check_height = df_2['H'][check_row]
                df_2['block_number'][check_row] = block_idx

                for q in range(check_row+1, len(df_2)):
                    if check_height + df_2['H'][q] <= H_lim:
                        df_2['block_number'][q] = block_idx
                        check_height += df_2['H'][q]

                df_2.sort_values(by=['block_number', 'product_type'], ascending=[True, True], inplace=True)
                df_2.reset_index(inplace = True, drop = True)
                k = len(df_2.loc[df_2['block_number'] == block_idx])
                check_row += k
                block_idx += 1
                
                if check_row == len(df_2):
                    df_block_data = pd.concat([df_block_data, df_2])
                    df_block_data.reset_index(inplace = True, drop = True)
                    cnt = 1
                    
    df_block_std = pd.concat([df_block_std, df_block_data])
    df_block_std.reset_index(inplace = True, drop = True)
    
    display(df_block_std)
    return df_block_std
```

```python
def create_block_data(df_block_std):
    df_input_lst = []

    for i in range(3):
        df_1 = df_block_std[df_block_std['destination'] == i+1]
        block_count = df_1['block_number'].unique()
        
        for j in range(len(block_count)):
            block = []            
            df_2 = df_1[df_1['block_number'] == j+1]
            df_2.reset_index(inplace=True, drop=True)         
            block_height = df_2['H'].sum()
            block_list = df_2.values.tolist()
            block_box = []
            
            for k in range(len(df_2)):
                block_box.append(block_list[k][5])
            
            block = block_list[0].copy()
            block[2] = block_height
            block[5] = block_box        
            df_input_lst.append(block)
            
    df_data = pd.DataFrame(df_input_lst)
    df_data.columns = ['L','W','H','S','destination','box_index','product_type','block_number']
    
    display(df_data)
    return df_data
```

```python
def loading(df_data):
    df_block_total_lst = [[]* x for x in range(len(df_data))]
    df_arrange = df_data.copy()
    position1_x.clear()
    position1_y.clear()
    position1_z.clear()
    position2_x.clear()
    position2_y.clear()
    position2_z.clear()
    truck_info.clear()
    mod_row_lst.clear()
    
    p = 0
    truck_cnt = 0
    mid_point = 0 
    while p < 1:
        sum_w = 0
        count = 0
        X = 0

        for i in range(len(df_arrange)):
            if sum_w + df_arrange['W'][i] <= W_truck:
                sum_w = sum_w + df_arrange['W'][i]
                count = count + 1
                
                if i+1 == len(df_arrange):
                    row_count.append(count)

            else:
                sum_w = 0
                row_count.append(count-1)
                sum_w = sum_w + df_arrange['W'][i]
                count = count + 1
                
        size = [[] for i in range(len(df_arrange))]
        for i in range(len(df_arrange)):
            size[i].append(df_arrange['L'][i])
            size[i].append(df_arrange['W'][i])
            size[i].append(df_arrange['H'][i])

        start = np.zeros((len(size),3), dtype='int64')
        end = np.zeros((len(size),3), dtype='int64')
        candidate = np.zeros((len(size),3,3), dtype='int64')
        check_length = []
        
        row = 0
        mod_row = 0
        for i in range(len(df_arrange)):

            for j in range(3):
                if j == 0:
                    candidate[i][j][0] = start[i][0] + size[i][0]
                    candidate[i][j][1] = start[i][1] 
                    candidate[i][j][2] = start[i][2]
                if j == 1:
                    candidate[i][j][0] = start[i][0]
                    candidate[i][j][1] = start[i][1] + size[i][1]
                    candidate[i][j][2] = start[i][2]
                if j == 2:
                    candidate[i][j][0] = start[i][0]
                    candidate[i][j][1] = start[i][1]
                    candidate[i][j][2] = start[i][2] + size[i][2]

            for j in range(3):
                end[i][j] = start[i][j] + size[i][j]

            if i < row_count[row+1]:

                if i+1 != len(df_arrange):
                    for j in range(3):
                        start[i+1][j] = candidate[i][1][j]

            elif i == row_count[row+1]:

                if i+1 != len(df_arrange):
                    check_length.clear()
                    
                    for j in range(row_count[row],row_count[row+1]):
                        check_length.append(candidate[j][0][0])
                        
                    x = max(check_length)
                    
                    start[i+1][0] = x
                    start[i+1][1] = 0
                    start[i+1][2] = 0
                    
                    row = row + 1

            if end[i][0] <= L_truck:
                
                if i == 0:
                    pass

                else:
                    if start[i][0] == start[i-1][0]:
                        pass
                    
                    else:
                        mod_row += 1

                mid_point = mid_point + 1

                position1_x.append(start[i][0])
                position1_y.append(start[i][1])
                position1_z.append(start[i][2])

                position2_x.append(end[i][0])
                position2_y.append(end[i][1])
                position2_z.append(end[i][2])

                truck_info.append(truck_cnt + 1)
                mod_row_lst.append(mod_row)


                continue

            else:

                df_arrange = df_arrange.loc[i:].copy()
                df_arrange = df_arrange.reset_index()
                df_arrange.drop(['index'],axis=1, inplace=True)

                break

        item_idx = mid_point
        truck_cnt = truck_cnt + 1

        if len(df_data) == item_idx :
            p = 1

    for i in range(len(df_data)):
        df_block_total_lst[i].append(df_data['L'][i])
        df_block_total_lst[i].append(df_data['W'][i])
        df_block_total_lst[i].append(df_data['H'][i])
        df_block_total_lst[i].append(df_data['S'][i])
        df_block_total_lst[i].append(df_data['destination'][i])
        df_block_total_lst[i].append(df_data['box_index'][i])
        df_block_total_lst[i].append(df_data['product_type'][i])
        df_block_total_lst[i].append(df_data['block_number'][i])
    
    df_block_total = pd.DataFrame(df_block_total_lst)
    df_block_total.columns = ['L', 'W', 'H', 'S', 'destination', 'box_index', 'product_type', 'block_number']
        
    df_block_total['truck_info'] = truck_info

    df_block_total['x1'] = position1_x
    df_block_total['y1'] = position1_y
    df_block_total['z1'] = position1_z
    df_block_total['x2'] = position2_x
    df_block_total['y2'] = position2_y
    df_block_total['z2'] = position2_z
    df_block_total['mod_row'] = mod_row_lst

    return df_block_total
```

```python
def box_data(df_block_total):
    
    df_total_raw_box = pd.DataFrame()
    df_raw_box = pd.DataFrame()
        
    for i in range(3):
        df_1 = df_block_std[df_block_std['destination'] == i+1]

        df_2 = df_block_total[df_block_total['destination'] == i+1]
        df_2.reset_index(inplace=True, drop=True)
        
        for j in range(len(df_2)):
            df_3 = df_1[df_1['block_number'] == j+1]
            df_3.reset_index(inplace=True, drop=True)
            
            df_3['truck_info'] = df_2['truck_info'][j]
            df_3['mod_row'] = df_2['mod_row'][j]
            
            for k in range(len(df_3)):
                if k == 0:
                    df_3['x1'] = df_2['x1'][j]
                    df_3['y1'] = df_2['y1'][j]
                    df_3['z1'] = df_2['z1'][j]
                    
                    df_3['x2'] = df_3['x1'] + df_3['L']
                    df_3['y2'] = df_3['y1'] + df_3['W']
                    df_3['z2'] = df_3['z1'] + df_3['H']
                
                else:
                    df_3['x1'][k] = df_2['x1'][j]
                    df_3['y1'][k] = df_2['y1'][j]
                    df_3['z1'][k] = df_3['z1'][k-1] + df_3['H'][k-1]

                    df_3['x2'][k] = df_3['x1'][k] + df_3['L'][k]
                    df_3['y2'][k] = df_3['y1'][k] + df_3['W'][k]
                    df_3['z2'][k] = df_3['z1'][k] + df_3['H'][k]

            df_raw_box = pd.concat([df_raw_box, df_3])
            df_raw_box.reset_index(inplace=True, drop=True)
    
    df_total_raw_box = pd.concat([df_total_raw_box, df_raw_box])
    
    display(df_total_raw_box)
    
    return df_total_raw_box
```
